### Basic Plots

```python
# importing required modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 50)
print (x[:10])
sinus = np.sin(x)
print (sinus[:10])

plt.plot(x, sinus)
plt.show()
```

```python
cosinus = np.cos(x)
plt.plot(x, cosinus)
plt.show()
```

```python
cosinus = np.cos(x)
plt.plot(x, cosinus, "o")
plt.show()
```

```python
cosinus = np.cos(x)
plt.plot(x, cosinus, "*")
plt.show()
```

```python
cosinus = np.cos(x)
plt.plot(x, cosinus, "--")
plt.show()
```

```python
cosinus = np.cos(x)
plt.plot(x, cosinus, "-")
plt.show()
```
